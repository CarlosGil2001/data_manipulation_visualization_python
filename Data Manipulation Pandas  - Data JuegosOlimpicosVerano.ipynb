```python
#Agregamos y reiniciamos el índice 0,1...n index, y creamos un nuevo dataframe con el índice
dt1=dt.reset_index()
```

```python
# Mostrar la cantidad de hombres y mujeres

#Primero agrupamos la columna Gender(género)
groupby_gender= dt1.groupby('Gender').count()[['index']] #Agrupamos por género y los contamos, y mostramos solo la columna 'index'
groupby_gender
```

```python
# GROUP BY - Agrupaciones

# Mostrar la cantidad de cada medalla (Gold, Silver, Bronce)
# Primero agrupamos por Medalla
dt1.groupby('Medal').count()[['index']] #Contamos por cada medalla y mostramos la columna index
```

```python
# Mostrar los 5 "Country" con más atletas

# Primero, agrupamos por "Country"
# Segundo, contamos los atletas por la agrupación
# Tercero, ordenamos de forma descendente la cantidad por agrupación
# Cuarto, mostramos solo los primeros 5
gr1=dt1.groupby('Country').count().sort_values('index',ascending=False).head(5)[['index']]
```

```python
# Mostrar la cantidad de athletas Hombres

# Primero, agrupamos por sexo
# Segundo, contamos por la agrupación
# Tercero, mostramos "index" y solo los de valores "Men"
dt1.groupby('Gender').count()['index']['Men']
```

```python
#Crear un dataframe a través de dt1 con columnas específicas

# Primero, creamos un DataFrame con el uso de pd (pandas)
# Segundo, de dt1 tomamos las columnas para el nuevo DataFrame
df2=pd.DataFrame(dt1, columns=['index','Year','Country','Gender','Medal'])
#Mostramos el nuevo dataframe df2
df2.head(10)
```

```python
# Mostrar los registros que tienen como país CHN

#df2[df2.Country=='CHN']
df2[df2['Country']=='CHN'] #Registros de df2 con valor CHN en country
```

```python
# UNIQUE - Datos únicos

# Mostrar los países
pd.unique(df2['Country'])
```

```python
# Mostrar valores de Medal
pd.unique(df2['Medal'])
```

```python
# Mostrar la cantidad de valores diferentes en la columna "Medal"
df2['Medal'].nunique()
```

```python
# Contar cuanto se repiten cada uno de los valores de la columna "Country"
# En otras palabras, cuantos atletas por países hay
df2['Country'].value_counts()
```

```python
# Contar cuanto se repiten cada uno de los valores de la columna "Year"
# En otras palabras, cuantos atletas participaron por "Year"
df2['Year'].value_counts()
```
