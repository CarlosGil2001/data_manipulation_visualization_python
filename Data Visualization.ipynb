```python
# Visualización de Datos con PANDAS
# Importar las librerías necesarias
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# DATA PARA GRAFICAR
dtsales=pd.read_csv('D:/data/supermarket_sales.csv') # leemos el archivo csv para importar la data
dtsales.head(10) # mostramos las primeras 10 filas
```

```python
# IMPORTAMOS NUEVA DATA PARA GRAFICAR
dtexam=pd.read_csv('D:/data/exams.csv') 
```

```python
# GRÁFICO DE BARRAS VERTICAL "bar" y HORIZONTAL "barh"

# Aregamos un índex para la mejor manipulación
dtexam=dtexam.reset_index() #Agregamos un índice 1,2,3,4,5...
```

```python
# GRÁFICO DE BARRAS VERTICAL "bar" y HORIZONTAL "barh"

#Estilo definido para el gráfico
plt.style.use('ggplot') # estilos de color rojo

# Cantidad de alumnos por raza/entidad y Nivel de educación de los padres
dtexam.pivot_table('index', index='race/ethnicity', columns='parental level of education', aggfunc=[np.count_nonzero]).plot(kind='bar', legend='Reverse')

#Título del gráfico
plt.title('Cantidad de Alumnos por Raza/Entidad y Educación de Padres')
#Título del eje x
plt.xlabel('Raz/Entidad')
#Título del eje y
plt.ylabel('Cantidad Alumnos')
```

```python
# GRÁFICO DE BARRAS VERTICAL "bar" y HORIZONTAL "barh"

#Estilo definido para el gráfico
plt.style.use('bmh') # estilos de color azul

#Cantidad de alumnos por "Género" y "test preparation course", que tienen una Nota en matemáticas mayor a 70 
# Primero, definimos el filtro math score>70
dtexam[dtexam['math score']>70].pivot_table('index', index='gender', columns='test preparation course', aggfunc=[np.count_nonzero]).plot(kind='bar', legend='Reverse',stacked=True)

#Título del gráfico
plt.title('Cantidad de Alumnos por Género y Condición Test')
#Título del eje x
plt.xlabel('Género')
#Título del eje y
plt.ylabel('Cantidad Alumnos si puntuación math > 70')
```

```python
# AÑADIMOS UNA NUEVA COLUMNA PARA CALCULAR EL TOTAL DEL PUNTAJE (math score + reading score + writing score)
dtexam['total score']=dtexam['math score'] + dtexam['reading score'] + dtexam['writing score']
```

```python
# GRÁFICO DE BARRAS VERTICAL "bar" y HORIZONTAL "barh"

#Cantidad de alumnos por Genero y Raz/Entidad de acuerdo a la Condición, que tinen un Total de puntación >=200
dtexam[dtexam['total score']>=200].pivot_table(['index'], index=['race/ethnicity', 'gender'], columns='test preparation course', aggfunc=[np.count_nonzero]).plot(kind='barh', legend='Reverse')

#Título del gráfico
plt.title('Alumnos por Grupo y Genero de acuerdo a la Condición')
#Título del eje x
plt.xlabel('Cantidad Alumnos si Total score > 200')
#Título del eje y
plt.ylabel('Raza/Entidad - Género')
```

```python
# GRÁFICO DE BARRAS VERTICAL "bar" y HORIZONTAL "barh"

#Definimos el estilo del gráfico
plt.style.use('classic') # estilo de color clasico

#Promedio del total score por Nivel de educación de padres
dtexam.pivot_table('total score', index='parental level of education', aggfunc=[np.mean]).plot(kind='bar', legend='Reverse')

#Título del gráfico
plt.title('Promedio del total score por Nivel de Educación de Padres')
#Título del eje x
plt.xlabel('Nivel de Educación de Padres')
#Título del eje y
plt.ylabel('Promedio de Total score')
```

```python
#Importar data para graficar
dtcovid=pd.read_csv('D:/data/covid-data.csv')
```

```python
# GRÁFICO DE BARRAS VERTICAL "bar" y HORIZONTAL "barh"

# AGREGAMOS NUEVAS COLUMNAS "year" y "month", PARA SU TRATAMIENTO POSTERIOR
import datetime as dt # importamos la libería para trabajar con fechas

 #Convertimos la columna date en tipo de dato datetime, ya que estaba con object
dtcovid['date']=dtcovid['date'].astype('datetime64')

# Creamos 2 nuevas columnas year y month
dtcovid['year']=dtcovid['date'].dt.year #Extraemos el year de la columna "date"
dtcovid['month']=dtcovid['date'].dt.month #Extraemos el month de la columna "date"
```

```python
# GRÁFICO DE BARRAS VERTICAL "bar" y HORIZONTAL "barh"

# Estilo para el gráfico
plt.style.use('ggplot') # estilos de color azul

# Total de fallecidos por Año y Mes
dtcovid.pivot_table(['total_deaths'], index='year', columns='month', aggfunc=[np.count_nonzero]).plot(kind='bar')

#Título del gráfico
plt.title('Total de Fallecidos Durante la Pandemia')
#Título del eje x
plt.xlabel('Año - Mes')
#Título del eje y
plt.ylabel('Cantidad Fallecidos')
```

```python
# GRÁFICO DE TORTA "pie"

# Estilo para el gráfico
plt.style.use('bmh') # estilos de color azul

# Total de casos por Continente
# subplots: es necesario habilitarlo para este tipo de gráficos
# cmap: brinda colores diferenciados en el gráfico
# figsize=(x,y): tamaño del gráfico
dtcovid.groupby('continent')[['total_cases']].count().plot(kind='pie', subplots=True, cmap='Paired')

#Título gráficos
plt.title('Total Casos por Continente')
#Dimensionamiento correcto de las parcelas
plt.axis('equal')
#Título del eje y
plt.ylabel('Total casos')
```

```python
# GRÁFICO DE TORTA "pie"

# Cantidad de Nuevos Casos de cada mes durante el 2020
#Primero, definimos el filtro de year=2020
dtcovid[dtcovid['year']==2020].pivot_table('new_cases', index='month', columns='year', aggfunc=[np.count_nonzero]).fillna(0).plot(kind='pie', subplots=True)

#Título gráficos
plt.title('Nuevos Casos por Mes del 2020')
#Dimensionamiento correcto de las parcelas
plt.axis('equal')
#Título del eje y
plt.ylabel('Meses del 2020')
```

```python
# GRÁFICO DE TORTA "pie"

#Definimos el estilo del gráfico
plt.style.use('classic') # estilo de color clasico

#Cantidad de Nuevos fallecidos en los primeros 6 meses de cada Año

#Primeros 6 meses de cada año
first_6_month=(dtcovid['month']>=1) & (dtcovid['month']<=6)
#Colocamos la variable de los 6 primeros meses, para filtrar
dtcovid[(first_6_month)].pivot_table('new_deaths', index=['year'], aggfunc=[np.count_nonzero]).plot(kind='pie', subplots=True)

#Título del gráfico
plt.title('Nuevos Casos de Fallecidos en los Primero 6 Meses de cada Año')
#Dimensionamiento correcto de las parcelas
plt.axis('equal')
#Título del eje y
plt.ylabel('Cantidad Nuevos Fallecidos')
```

```python
# GRÁFICO DE TORTA "pie"

#Definimos el estilo del gráfico
plt.style.use('bmh')

# % Promedio de New Test de las localidades de Oceania con una población mayor a 1 000 000

# Condición: Que continente sea Oceanía que una población mayor a 1 mill
filtro=(dtcovid['continent']=='Oceania') & (dtcovid['population']>1000000)
#Colocamos el filtro
dtcovid[filtro].pivot_table(['new_tests'], index='location', aggfunc=[np.mean]).plot(kind='pie', subplots=True, autopct='%1.1f%%', shadow=True, startangle=180)

#Título del gráfico
plt.title('Promedio de Nuevos Test por País de Ocenía con Población > 1 mill')
#Dimensionamiento correcto de las parcelas
plt.axis('equal')
#Título del eje x
plt.xlabel('País')
#Título del eje y
plt.ylabel('Nuevos test')
```

```python
# GRÁFICO DE TORTA "pie"

#Definimos el estilo del gráfico
plt.style.use('ggplot')

# Promedio de tasa positiva y test por caso durante el año 2022 y 2020

#Filtro para el año 2020
data_2020=(dtcovid['year']==2020) | (dtcovid['year']==2022)
#Agregamos el filtro para el DataFrame
dtcovid[data_2020].groupby('year')[['positive_rate','tests_per_case']].mean().plot(kind='pie', subplots=True, figsize=(10,10))
```

```python
# AREA - "area" - PARA DATOS CUANTITATIVOS

#Definimos el estilo del gráfico
plt.style.use('bmh')

#Es coveniente agrupar los valores para utilziar el eje x
# stacked=False: suavizar el color

#Mostrar la cantidad de nuevos casos y total de casos por Continente
dtcovid.groupby('continent').count().plot(kind='area', y=['total_cases','new_cases'])

#Título del gráfico
plt.title('Total de Casos y Nuevos Casos por Continente')
#Título del eje x
plt.xlabel('Continente')
#Título del eje y
plt.ylabel('Total Casos - Nuevos Casos')
```

```python
# AREA - "area" - PARA DATOS CUANTITATIVOS

#Definimos el estilo del gráfico
plt.style.use('ggplot')

#Motsrar el el Total de test en los meses del año 2021

#Para los meses del año 2021
year_2021=dtcovid['year']==2021
#Aplicamos la variable del filtro al DataFrame
dtcovid[year_2021].pivot_table('total_tests', index='month', aggfunc=[np.count_nonzero]).plot(kind='area')

#Título del gráfico
plt.title('Total de Test en el 2021')
#Título del eje x
plt.xlabel('Meses')
#Título del eje y
plt.ylabel('Total Test')
```

```python
# AREA - "area" - PARA DATOS CUANTITATIVOS

#Definimos el estilo del gráfico
plt.style.use('bmh')

#Promedio de Nuevas Pruebas, Total de vacunaciones, Nuevas vacunas por iso_code en los primeros 4 meses de los años 2020 y 2022

#Filtro
filtro=((dtcovid['year']==2020) | (dtcovid['year']==2022)) & (dtcovid['month']>=1) & (dtcovid['month']<=4)
#Ponemos el filtro realizado
dtcovid[filtro].groupby('iso_code').count().plot(kind='area', y=['new_tests','total_vaccinations','new_vaccinations'])

#Título del gráfico
plt.title('New Tests, Total and New Vaccinations')
#Título del eje x
plt.xlabel('Iso Code')
#Titulo del eje y
plt.ylabel('Quantity')
```

```python
# GRÁFICO DE LINEA "line"- Para medir la evolución de una o varias variables numéricas

#figsize(largo, ancho): tamaño del gráfico
#lw=1 : tamaño de las líneas
#marker='o' : mostrar punto en las lineas, para mejor visibilidad de evolución
#linestyle='dashed' : líneas : (- - - - -)
#markerfacecolor='red' : los puntos de colore rojo
#color='skyblue': color de las líneas

#Definimos el estilo del gráfico
plt.style.use('ggplot')

# Mostrar la evolución de la cantidad del total de casos por iso_code del continente de Sur América en el año 2020

#Filtro para el continente
filtro=(dtcovid['continent']=='South America') & (dtcovid['year']==2020)
#Asignar la variable del filtro
dtcovid[filtro].groupby('iso_code').count().plot(kind='line', y=['total_cases'],figsize=(13,5),lw=1,marker='o')

#Título del gráfico
plt.title('Total de Casos en South Amerca en 2020')
#Agregar legend
plt.legend()
#Título del eje x
plt.xlabel('Iso Code')
#Título del eje y
plt.ylabel('Total Casos')
```

```python
# GRÁFICO DE LINEA "line"- Para medir la evolución de una o varias variables numéricas

#Definimos el estilo del gráfico
plt.style.use('bmh')

# Mostrar la evolución del promedio de Nuevos casos y Total de fallecidos por Continente en el primer mes de Pandemia

#Filtro para el primer mes de pandemia
year_2020=(dtcovid['year']==2020) & (dtcovid['month']==3)

dtcovid[year_2020].pivot_table(['new_cases','total_deaths'], index=['continent'], aggfunc=[np.mean]).plot(kind='line',figsize=(12,4),marker='o')

#Título del gráfico
plt.title('Nuevos Casos y Total de Fallecidos en el 1er Mes de Pandemia')
#Agregar legend
plt.legend()
#Título del eje x
plt.xlabel('Continente')
#Título del eje y
plt.ylabel('Cantidad')
```

```python
# GRÁFICO DE LINEA "line"- Para medir la evolución de una o varias variables numéricas

#Definimos el estilo del gráfico
plt.style.use('fivethirtyeight')

#Mostrar la evolución de Nuevos Casos, Total de Casos, Nuevos Fallecidos y Total de fallecidos por año
dtcovid.groupby('month').count()[['total_cases', 'total_deaths','total_tests','total_vaccinations']].plot(kind='line', marker='o', figsize=(14,6),markerfacecolor='orange')

#Título del gráfico
plt.title('Total de Casos, Fallecidos, Test y Vacunados por Mes')
#Agregar legend
plt.legend()
#Título del eje x
plt.xlabel('Mes')
#Título del eje y
plt.ylabel('Cantidad')
```

```python
# GRÁFICO DE DISPERSIÓN - muestra la relación entre 2 o más variables numéricas

#import seaborn as sns #importamos la librería
#lmplot: para correlaciones lineal

# c:indicar el color de los valroes
# cmap: indicar el mapa de colores

#Definimos el estilo del gráfico
plt.style.use('bmh')

# Mostrar la relación entre los Nuevos Casos y el Total de Casos en el mes de enero y julio del 2022

#Filtro
year_2022=(dtcovid['year']==2022) & (dtcovid['month'].isin([1,7]))
#Agregamos el filtro
dtcovid[year_2022].plot(kind='scatter', x='new_cases', y='total_cases')
#plot(x='total_cases', y='continent', data=dtcovid, fit_reg=False, hue='total_cases', lengend=False, palette='Paired')

#Título del gráfico
plt.title('Relación Nuevos Casos y Total de Casos')
#Título del eje x
plt.xlabel('Nuevos casos')
#Título del eje y
plt.ylabel('Total casos')
```

```python
# GRÁFICO DE DISPERSIÓN - muestra la relación entre 2 o más variables numéricas

#Motsrar la relación entre New Test, Total Test y Population por Continente de Africa

#Filtro para el continent
cont=dtcovid['continent']=='Africa'
#Agregamos el filtro
dtcovid[cont].plot(kind='scatter', x='total_tests',y='new_tests',c='population',cmap='coolwarm')

#Título del gráfico
plt.title('Relación Total Tests - Nuevos Tests')
#Título del eje x
plt.xlabel('Total tests')
#Título del eje y
plt.ylabel('Nuevos tests')
```
