```python
#IMPORTAMOS DE DATA CSV
import pandas as pd #importamos panda
import numpy as np # importamos numpy

# Leemos el csv en la ruta
dtcovid= pd.read_csv('D:/data/covid-data.csv')
dtcovid.head(10) #Mostramos los primeros 10 datos
```

```python
# AGREGAMOS UN ÍNDICE PARA LA MEJOR MANIPULACIÓN
dtcovid=dtcovid.reset_index() #Agregamos un índice 1,2,3,4,5...
```

```python
# MOSTRAR TOTAL DE CASOS POR "Continent"

# Primero, agrupamos por continente y contamos
# Segundo, ordenamos de forma desc y mostramos el valor index

#dtcovid['continent'].value_counts()
dtcovid.groupby('continent').count().sort_values('index',ascending=False)[['index']]
```

```python
# Remplazar los valores NULOS por 0
dtcovid.fillna(0)
```

```python
# CREAR LAS COLUMNAS year Y month

import datetime as dt # importamos la libería para trabajar con fechas

 #Convertimos la columna date en tipo de dato datetime, ya que estaba con object
dtcovid['date']=dtcovid['date'].astype('datetime64')

# Creamos 2 nuevas columnas year y month
dtcovid['year']=dtcovid['date'].dt.year #Extraemos el year de la columna "date"
dtcovid['month']=dtcovid['date'].dt.month #Extraemos el month de la columna "date"
# Verificamos
dtcovid 
```

```python
# LISTAR LA CANTIDAD DE MUERTOS POR AÑO DIVIDIDO POR CONTINENTE

# Primero, agrupamos por "year" y "continent"
# Segundo, contamos por la agrupación
# Tercero, mostramos la cantidad 
# unstack(): para mostrar en formato tabla
dtcovid.groupby(['year','continent'])[['total_deaths']].count().unstack()
```

```python
# USO DE PIVOT TABLES

# LISTAR LA CANTIDAD DE MUERTOS POR CONTINENTE DIVIDIDO POR AÑO

# 'total_deaths' : campo en el que se realizará la operación
# index: representa a la fila
# columns: representa la columna
# aggfunc: se especifica el tipo de operación que se realizará
# margin: mostrar una columna de 'all' que es el acumulado total

dtcovid.pivot_table('total_deaths', index=['continent'], columns='year', aggfunc=[np.count_nonzero],margins=True)
```

```python
# USO DE PIVOT TABLES

# LISTAR EL PROMEDIO DE NUEVOS CASOS POR AÑO Y MES DIVIDIO EN CONTINENTES
# fillna(0): reemplaza los valores NULOS por 0
dtcovid.pivot_table('new_cases',index=['year','month'], columns=('continent'), aggfunc=[np.mean], margins=True).fillna(0)
```

```python
# USO DE PIVOT TABLES

# VACUNAS TOTALES Y PACIENTES HOSPEDADOS POR CONTINENTE
#dtcovid.groupby('continent').sum()[['hosp_patients','total_vaccinations']].fillna(0)
dtcovid.pivot_table(['total_vaccinations','hosp_patients'],index='continent',aggfunc=[np.sum], margins=True).fillna(0)
```

```python
# USO DE PIVOT TABLES

# MOSTRAR LOS NUEVOS CASOS QUE SEAN MAYOR A 1000 DURANTE EL MES DE ENERO DE CADA AÑO
dtcovid.groupby(['year','month']).count()[['new_cases']].query('month==1 & new_cases')
```

```python
# CONCATENACIÓN - combina DataFrames

# Primero, creamos 2 dataframes
# DataFrame1
df1=pd.DataFrame({'A': ['A0','A1','A2','A3'],
                  'B': ['B0','B1','B2','B3'],
                  'C': ['C0','C1','C2','C3'],
                  'D': ['D0','D1','D2','D3']},
                    index=[1,2,3,4])

# DataFrame2
df2=pd.DataFrame({'A': ['E0','E1','E2','E3'],
                  'B': ['F0','F1','F2','F3'],
                  'C': ['F0','F1','F2','F3'],
                  'D': ['G0','G1','G2','G3']},
                    index=[1,2,3,4])
```

```python
# CONCATENACIÓN - combina DataFrames

# concat : concatena los dataframes que se le asigna
# axis=0 : concatenar verticalmente - predeterminado
# axis=1 : concatenar horizontalmente
pd.concat([df1,df2])
```

```python
# CONCATENACIÓN - combina DataFrames

# concatenar dataframe horizontalmente, los index coinciden, por eso no hay valores nulos
pd.concat([df1,df2], axis=1) # axis=1 para concatenación horizontal
```

```python
# FUSIÓN MERGE

# merge(): permite fusionar DataFrames utilizando la lógica parecida a la relación entre tabla

#Creamos los DataFrame dfpersona y dfpais
dfpersona= pd.DataFrame({'Nombre' : ['Carlos','Pedro','Lucia','Esther','Jose','Mariana'],
                          'Apellido': ['Gil','Saravia','Gómez','Espinoza','Velez','Tello'],
                          'Edad': [21,16,20,25,22,23],
                          'idPais': ['PER','ESP','CHL','ECU','ESP','CHL']})


dfpais= pd.DataFrame({'idPais' : ['PER','ESP','ECU','BRA','CHL','ARG','URU','COL'],
                           'Nombre' : ['Perú','España','Ecuador','Brasil','Chile','Argentina','Uruguay','Colombia']})
```

```python
# Verificamos la creación de los Dataframe
print('----------PERSONA--------')
print(dfpersona)

print('-----------PAÍS----------')
print(dfpais)
```

```python
# INNER JOIN - Intersección

#Sintáxis: pd.merge(dataframe1,dataframe2, how='relación', on='columna de relación')
# how='inner'
# how='outer'
# how='left'
# how='right'

# INNER JOIN
# Relacionamos los dataframe dfpersona y dfpais
pd.merge(dfpersona,dfpais, how='inner',on='idPais')
```

```python
# OUTER JOIN - Fusionada ambos dataframe por completo

# Relacionamos los dataframes por completo (outer)
# fillna('Sin dato'): poner el texto específicado a valores nulos 
pd.merge(dfpersona,dfpais, how='outer', on='idPais').fillna('Sin dato')
```

```python
# LEFT JOIN 
#Mostrar los registros de la primera tabla(izquierda) y los registros si incluso no coinciden con la segunda tabla(derecha)

pd.merge(dfpersona, dfpais, how='left', on='idPais')
```

```python
# RIGHT JOIN
# Mostar los valores de la segunda tabla(derecha) y los registros si incluso no coinciden de la primera tabla(izquierda)
pd.merge(dfpersona,dfpais, how='right', on='idPais')
```
