```python
#Pandas es una biblioteca de código abierto de Python que proporciona análisis y manipulación de datos.

# csv: valores separados por comas

# 1 PASO: Importar librería pandas
import pandas as pd
```

```python
# 2 PASO: Leer archivo
#Creamos una variable y leemos el dato
data= pd.read_csv('D:/forbes.csv')
#Para saber si se ha leído, mostrar los primeros 5 registros
data.head()
```

```python
#Mostrar las columnas del archivo csv
data.columns
```

```python
#Para ver todos los datos
data
```

```python
#Para ver las primeras 6 filas
print(data[:6])
```

```python
#Para ver solo una fila especifica
print(data.loc[0])  #Muestra el registro n° 0
```

```python
#NOTA: Si especificamos [['Company']] nos mostrará el campo y el valor
#      Si especificamos ['Company'] solo nos mostrará el valor

#Para mostrar solo las columnas especificadas de todos los datos
print(data[['Company','Sales']])  #Muestra las columnas "Company" y "Sales"
```

```python
#Para mostrar una columna en específico de solo de un registro
data.loc[0]['Company'] #Muestra la columna "Company" del registro 0
```

```python
#Para mostrar columnas especificadas de un registro
data.loc[1][['Company','Sales']]  #Muestra las columnas "Company" y "Sales" del registro 1
```

```python
#Crear una nueva columna a partir de la combinación de otras
data['Continent-Country']= data['Continent'] + ' ' + data['Country']  #Crea una nueva columna a partir de la combinación de "Continent" y "Country"
data[['Continent-Country']] #Mostrar la nueva columna creada
```

```python
#Eliminar columna
data.drop('Continent-Country',axis=1) # axis = 0 elimina filas(index) axis = 1 elimina columnas
```

```python
#Sumar todas las sales (ventas) - Columnas sales
sum(data['Sales']) # suma los valores de la columna sales
```

```python
#Redondear la suma de los valores de la columna sales (venta)
#round = redondea un valor
round(sum(data['Sales']))
```

```python
#Promedio de las columnas numéricas
data.mean()  #Muestra el promedio de las columnas con valor numérico
```

```python
#Promedio de una columna en especifico
data['Sales'].mean()
```

```python
#Promedio de columnas especificadas
data[['Sales','Rank']].mean()
```

```python
#Para mostrar estadísticas básicas (count, mean, mix, max, etc) de una columna
data['Sales'].describe()
```

```python
#Listar los registros que tengan una venta menor e igual a 0.5
data[data['Sales']<=0.5]
```

```python
#Listar los registro que tengan como "Country" a Japan, y que solo muestre la columna de "Country"
data[data['Country'] == 'Japan'][['Country']]
```

```python
#Listar sales(ventas) que sean mayores e iguales que 50 y menos que 90, y solo mostrar las columnas "Company" y "Sector"
data[(data['Sales']>=50) & (data['Sales']<90)][['Company','Sector']]
```

```python
#Uso de .query() - Busca condicional

#Lista de sales(venta) mayores a 20
# data[data['Sales']>20]
data.query('Sales > 20')
```

```python
#Lista de Companys que tienen el Assets(activos) mayores que 20 y menores e iguales que 130
data.query('Assets>20 and Assets<=130')[['Company']] #Mostramos solo la columna "Company"
```

```python
#Agregamos un índice
data1=data.reset_index()
data1
```

```python
# Suma de ventas por Company
# Primero agrupamos por "Company"
# Segundo, sumamos las ventas por la agrupación de "Company"
# Tercero, solo mostramos la columna de "Sales"
data1.groupby('Company').sum('Sales')[['Sales']]
```

```python
# Cantidad de ventas por "Sector"
data1.groupby('Sector').count()[['Sales']]
```
